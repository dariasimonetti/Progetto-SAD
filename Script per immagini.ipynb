```python
import os
import nibabel as nib
import numpy as np
import imageio
```

```python
# Cartella principale contenente "Patient-1" ... "Patient-60"
root_dir = "D:/Daria/Desktop/DATASET-SLA-DARIA/8bctsm8jz7-1 (1)/8bctsm8jz7-1/Brain MRI Dataset of Multiple Sclerosis with Consensus Manual Lesion Segmentation and Patient Meta Information"

# Cartella dove salvare le immagini PNG
output_dir = "D:/Daria/Desktop/DATASET-SLA-DARIA/PNG_Output"

os.makedirs(output_dir, exist_ok=True)

# Lista dei template dei nomi dei file da cercare
nii_files = [
    "Flair.nii",
    "LesionSeg-Flair.nii",
    "LesionSeg-T1.nii",
    "LesionSeg-T2.nii",
    "T1.nii",
    "T2.nii"
]
```

```python
for patient_folder in sorted(os.listdir(root_dir)):
    patient_path = os.path.join(root_dir, patient_folder)
    if not os.path.isdir(patient_path):
        continue

    print(f"Processando {patient_folder}...")

    patient_output = os.path.join(output_dir, patient_folder)
    os.makedirs(patient_output, exist_ok=True)

    for nii_file in nii_files:
        patient_id = patient_folder.split("-")[1]
        file_name = f"{patient_id}-{nii_file}"
        nii_path = os.path.join(patient_path, file_name)

        if not os.path.exists(nii_path):
            print(f" File non trovato: {nii_path}")
            continue

        scan_name = nii_file.replace(".nii", "")
        scan_output = os.path.join(patient_output, scan_name)
        os.makedirs(scan_output, exist_ok=True)

        # Carica il file NIfTI
        nii_img = nib.load(nii_path)
        img_data = nii_img.get_fdata()

        # Normalizzazione
        if "LesionSeg" not in scan_name:
            # immagini mediche normali
            img_data = (img_data - np.min(img_data)) / (np.max(img_data) - np.min(img_data)) * 255
        else:
            # maschere LesionSeg → 0/1 → 0/255 per renderle visibili
            img_data = (img_data * 255)

        img_data = img_data.astype(np.uint8)

        # Salva slice come PNG
        for i in range(img_data.shape[2]):
            slice_img = img_data[:, :, i]
            slice_filename = os.path.join(scan_output, f"{scan_name}_slice_{i:03d}.png")
            imageio.imwrite(slice_filename, slice_img)

print("Conversione completata!")
```

```python
import os
import nibabel as nib
import numpy as np
import pandas as pd
from PIL import Image
```

```python
base_dir = "D:/Daria/Desktop/DATASET-SLA-DARIA/8bctsm8jz7-1 (1)/8bctsm8jz7-1/Brain MRI Dataset of Multiple Sclerosis with Consensus Manual Lesion Segmentation and Patient Meta Information"
output_dir = "D:/Daria/Desktop/DATASET-SLA-DARIA/PNG_Output_prof1"
excel_file = base_dir+"/Supplementary Table 1 for patient info .xlsx"
```

```python
df = pd.read_excel(excel_file, header=1)
print("Colonne trovate:", df.columns.tolist())

# Crea un dizionario ID (int) → EDSS
edss_dict = dict(zip(df["ID"].astype(int), df["EDSS"]))

# Assicurati che la cartella di output esista
os.makedirs(output_dir, exist_ok=True)
```

```python
def save_slices(nifti_path, modality, patient_id, edss):
    img = nib.load(nifti_path)
    data = img.get_fdata()
    
    for i in range(data.shape[2]): 
        slice_2d = data[:, :, i]
        slice_norm = 255 * ((slice_2d - np.min(slice_2d)) / (np.ptp(slice_2d)))
        slice_img = Image.fromarray(slice_norm.astype(np.uint8))
        slice_path = os.path.join(output_dir, f"{patient_id}_{modality}_slice{i}_EDSS{edss}.png")
        slice_img.save(slice_path)
```

```python
for patient_folder in os.listdir(base_dir):
    if not patient_folder.startswith("Patient-"):
        continue

    patient_id = patient_folder.split("-")[-1]
    if not patient_id.isdigit():
        continue

    patient_id = int(patient_id)
    edss = edss_dict.get(patient_id)

    if edss is None:
        print(f"EDSS non trovato per paziente {patient_id}")
        exit() #qusto non dovrebbe mai accadere... 
        continue

    patient_path = os.path.join(base_dir, patient_folder)

    for file in os.listdir(patient_path):
        if "LesionSeg" in file: #Ignoriamo le immagini segmentate per questo progetto...
            continue
        if file.endswith(".nii") and ("Flair" in file or "T1" in file or "T2" in file):
            
            if "Flair" in file:
                modality = "Flair"
            elif "T1" in file:
                modality = "T1"
            elif "T2" in file:
                modality = "T2"
            #TODO: Qui andrebbe controllato se sono disponibili anche altri layer (es T3)...
            
                
            full_path = os.path.join(patient_path, file)
            save_slices(full_path, modality, patient_id, edss)
```

```python
def save_merged_slice(nifti_path, modality, patient_id, edss):
    # === 1) Definisci la cartella di output ===
    output_dir = "D:/Daria/Desktop/DATASET-SLA-DARIA/PNG_Output_prof2"
    os.makedirs(output_dir, exist_ok=True)  # ← CREA LA CARTELLA SE NON ESISTE

    # === 2) Carica la NIfTI e unisci le slice ===
    img = nib.load(nifti_path)
    data = img.get_fdata()

    merged_slice = np.sum(data, axis=2)

    # === 3) Normalizzazione 0-255 ===
    if np.ptp(merged_slice) == 0:
        merged_norm = np.zeros_like(merged_slice)
    else:
        merged_norm = 255 * ((merged_slice - np.min(merged_slice)) / np.ptp(merged_slice))

    merged_img = Image.fromarray(merged_norm.astype(np.uint8))

    # === 4) Costruisci nome file output ===
    file_name = f"{patient_id}_{modality}_merged_EDSS{edss}.png"
    merged_path = os.path.join(output_dir, file_name)

    # === 5) Salva ===
    merged_img.save(merged_path)
    print(f"✔ Salvata: {merged_path}")
```

```python
import os
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from scipy.ndimage import label, binary_opening, binary_closing
import pandas as pd

# Cartella con le slice
slice_dir = "D:/Daria/Desktop/DATASET-SLA-DARIA/PNG_Output_prof1"
output_dir = "D:/Daria/Desktop/DATASET-SLA-DARIA/controllo_zone_bianche"
os.makedirs(output_dir, exist_ok=True)

# Soglia e parametri
THRESHOLD = 185
BORDER = 10       # Ignora 5 pixel dai bordi
MIN_SIZE = 20    # Ignora regioni più piccole di 20 pixel

slice_files = [f for f in os.listdir(slice_dir) if f.endswith(".png")]

results = {}

for f in slice_files:
    img_path = os.path.join(slice_dir, f)
    img = Image.open(img_path).convert("L")
    img_array = np.array(img)
    
    # Taglio i bordi
    img_cropped = img_array[BORDER:-BORDER, BORDER:-BORDER]
    
    # Maschera binaria
    binary_mask = img_cropped > THRESHOLD
    
    # Filtri morfologici per pulire il rumore
    binary_mask = binary_opening(binary_mask, structure=np.ones((3,3)))
    binary_mask = binary_closing(binary_mask, structure=np.ones((3,3)))
    
    # Labeling delle componenti connesse
    labeled_array, num_features = label(binary_mask)
    
    # Rimuovo regioni troppo piccole
    sizes = np.bincount(labeled_array.ravel())
    mask_clean = np.zeros_like(binary_mask)
    for i, size in enumerate(sizes):
        if size > MIN_SIZE and i != 0:
            mask_clean[labeled_array == i] = 1
    
    # Conta le zone bianche finali
    labeled_final, num_zones = label(mask_clean)
    
    # Estrai info da filename
    parts = f.split("_")
    patient_id = int(parts[0])
    modality = parts[1]
    
    # Salva maschera sovrapposta per controllo
    overlay = np.stack([img_cropped]*3, axis=-1)  # RGB
    overlay[mask_clean==1] = [255,0,0]           # Rosso per zone bianche
    plt.imsave(os.path.join(output_dir, f"{f}_overlay.png"), overlay)
    
    # Salva dati
    if patient_id not in results:
        results[patient_id] = {"Flair": [], "T1": [], "T2": []}
    results[patient_id][modality].append(num_zones)

# Preparo DataFrame per Excel
data_for_excel = []
for pid, modalities in results.items():
    row = {
        "ID": pid,
        "FLAIR_white_count": np.mean(modalities["Flair"]),
        "T1_white_count": np.mean(modalities["T1"]),
        "T2_white_count": np.mean(modalities["T2"])
    }
    data_for_excel.append(row)

df_out = pd.DataFrame(data_for_excel)
df_out.to_excel("zone_bianche_per_paziente.xlsx", index=False)
print("Fatto! Maschere salvate e dataset aggiornato.")
```
